```python
from sklearn.model_selection import train_test_split
#Splitting the test and train data sets as processing is done
p=x.iloc[113936:]
x=x.iloc[:113936]
y=m.iloc[:,-1]
print(len(y))
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
print(x)
m
```

```python
from sklearn.ensemble import RandomForestClassifier
#Building the model
model=RandomForestClassifier(criterion="entropy",n_estimators=500,n_jobs=80 )
# model1=SVC(kernel="rbf")
print(2)
model.fit(x,y)
x['USERZIPCODE'].fillna(x_train.USERZIPCODE.mode()[0],inplace=True)
x.info()
model.fit(x,y)
```

```python
#predicting the score
from sklearn.metrics import accuracy_score
accuracy_score(y_test,model.predict(x_test))
#predicting the outputs of test data
k=model.predict(p)
```
